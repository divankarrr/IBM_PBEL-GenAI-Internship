```python
def create_diamond(edge_length):
    if edge_length <= 0:
        print("Edge length must be a positive integer.")
        return
    # Upper part of the diamond
    for i in range(1, edge_length + 1):
        print(" " * (edge_length - i) + "* " * i)
    # Lower part of the diamond
    for i in range(edge_length - 1, 0, -1):
        print(" " * (edge_length - i) + "* " * i)
```

```python
create_diamond(5)
```

```python
create_diamond(0)
create_diamond(-1)
create_diamond(8)
create_diamond(12)
```

```python
def reverse_string(str1):
  rev = ""
  for ch in str1:
    rev = ch + rev
  return rev
```

```python
class stringutils:
  def __init__(self,string):
    self.string = string
  def palindrom(self):
    return self.string == self.string[::-1]
  def srtlength(self):
    return len(self.string)

obj = stringutils("HappyppaH")
print(obj.palindrom())
print(obj.srtlength())
obj.string
```
