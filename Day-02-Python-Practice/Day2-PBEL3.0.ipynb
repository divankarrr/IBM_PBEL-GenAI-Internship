1. **Make a function to print a hollow square pattern.**

   * Take the side length as an argument.
   * Example for length = 5:

   ```
   *****
   *   *
   *   *
   *   *
   *****
   ```

````python
def pattern(length):
  for i in range(length):
    for j in range(length):
      if(i==0 or i==length-1 or j==0 or j==length-1):
        print("*",end="")
      else:
        print(" ",end="")
    print()
pattern(5)
````

3. **Make a function to count vowels in a string.**

   * Take a string as an argument.
   * Return the total number of vowels.

````python
def vowels(word):
  c=0
  for i in word:
    if(i=='a' or i=='e' or i=='i' or i=='o' or i=='u' or i=='A' or i=='E' or i=='I' or i=='O' or i=='U'):
      c+=1
  return c
print(vowels("hello divankar"))
````

5. **Make a function to find the largest number in a list without using `max()`.**

   * Take a list as an argument.
   * Return the largest value.

````python
def maxnum(list):
  max=list[0]
  for i in list:
    if(i>max):
      max=i
  return max
print(maxnum([1,2,30,4,5,6,7,80,9,10]))
````

7. **Make a function to check whether a number is prime.**

   * Take a number as an argument.
   * Return `True` or `False`.

````python
def isprime(num):
  for i in range(2,num):
    if(num%i==0):
      return False
  return True
print(isprime(10))
````

9. **Make a function to calculate the factorial of a number using a loop.**

   * Take a number as an argument.
   * Do not use recursion or built-in libraries.

````python
def factorial(n):
    result = 1

    for i in range(1, n + 1):
        result *= i

    return result

num = int(input("Enter a number: "))
print("Factorial =", factorial(num))
````

10. **Make a class called `Student` with the following functions:**

    * Add student name and marks.
    * Display student details.
    * Check whether the student passed (marks ≥ 40).

````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display_details(self):
        print("Student Name:", self.name)
        print("Marks:", self.marks)

    def check_pass(self):
        if self.marks >= 40:
            print("Result: Pass")
        else:
            print("Result: Fail")


# Input
name = input("Enter student name: ")
marks = int(input("Enter marks: "))

# Create object
student = Student(name, marks)

# Display details
student.display_details()

# Check pass/fail
student.check_pass()
````
