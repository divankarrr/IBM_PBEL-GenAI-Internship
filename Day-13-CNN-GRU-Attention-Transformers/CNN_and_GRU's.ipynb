**CNN(Convolutional Neural Network)**

```python
# What is CNN.?
# CNN stands for CONVOLUTIONAL NEURAL NETWORKS.
# It is a part of deep learning.
# It is used to process large volume of data.
# Used to process images and extract features from those images.
# Features of images can be edges, shapes, and texture.

# Real Life analogy of CNN:
# It works similar to human identyfing an image
# First it detects the edges.
# It checks the shape of the image.
# It helps in configuring between objects.
```

```python
# # Architecture of CNN
# # Steps about the working of CNN
# 1. First we take an input image (64x64x3)
# 2. Create an convolutional layer
# 3. ReLu Activation
# 4. Pooling layer
# 5. Flatten
# 6. Dense layer
# 7. Output
```

```python
# # Pooling
# # Pooling reduces the image size
# # It keeps the important features
# # There are two types of pooling:
# # 1. Max pooling -> It will take the highest value
# # 2. Average Pooling -> It will take the average value

# # Example:
# Consider there is an image. First the pixels are converted into matrix.
# Matrix:
#  [1,2]
#  [3,4]
# Max Pool -> 4
# Average Pool -> 2 or 3 (based on your learning pattern.)

```

```python
# Padding
# Padding adds extra pixels
# So that the output size is controlled
# Two types of padding are present:
# 1. Valid Padding: No padding.
# 2. Same Padding: Output size same.
```

```python
# Step 1: Importing Tools
import tensorflow as tf # tesnorflow is an AI library and consist of the full AI toolkit
from tensorflow.keras import layers, models # we are accessing keras and importing the layers and models
# You are opening a toolbox
# TensorFlow: Full AI kit
# layers and models: Tools to build the brain


# Step 2: Creating the brain
model = models.Sequential([
# Sequential means we are building the model layer by layer
# Like stacking the lego blocks

    # Step3: Looking in an image
    layers.Conv2D(32,(3,3),activation = 'relu',input_shape=(64,64,3)),
    # The model looks at a 64x64 color image
    # (3,3) = small window which scanning the image
    # 32 = it looks like 32 different patterns
    # relu helps in keeping useful info, removes noise

    # Step 4: Shrinking the image
    layers.MaxPooling2D((2,2)),
    # Reduces the image size
    # Keeps only the important parts


    # Step 5: Learning more patterns
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    # It looks deeper into the image
    # Find more complex things like shapes, object parts
    # Filtering it to 64 more parts for better understanding

    # Step 6: Flattening
    layers.Flatten(),
    # Converts image and its patterns into a long list of numbers
    # Turned this 2d image into a single row of numbers. Decision making becomes easier.

    # Step 7: Decision Making
    layers.Dense(64,activation='relu'),
    # It processses all features
    # Tries to understand patterns

    # Step 8: Final answer
    layers.Dense(10,activation='softmax')
    # Output layer: Model chooses between 10 categories
    # Example: Cat, dog, car etc
    # softmax: gives prob. example: '90% sure that it is a cat'
])

model.compile(optimizer='adam', # How the model improves
              loss = 'categorical_crossentropy', # Measure how wrong your prediction is
              metrics = ['accuracy']) # How many times your model is predicting correctly

model.summary()
```

**GRU(GATED RECURRENT UNIT)**

```python
# What is GRU.?
# To solve the issues which were present in RNN, we made the use of GRU.
# Problems in RNN (Recurrent Neural Network)
# RNN forgets long term information
# Vanishing gradient problem is persistent in RNN so as to solve that problem, we made use of GRU.

## Definition
# GRU stands for GATED RECURRENT UNIT.
# GRU is an updated version of RNN.
# GRU uses gates to control the memory.
# So with the help of these gates GRU helps you to learn the patterns for a long time.
```

```python
# # Architecture
# GRU uses two gates in its architecture:
# 1. Update Gate (z) -> How much to remember
# 2. Reset Gate (r) -> How much to forget

# Intution through human analogy
# Keeps important information
# Forgets useless information
```

```python
# Step 1: Creating the model
model = tf.keras.Sequential([
# We are building the model step by step like LEGO blocks


    # Step 2: Understanding words(Embedding layer)
    tf.keras.layers.Embedding(input_dim = 5000, output_dim = 64),
    # Converts wrods into numbers(because computers dont understand text)
    # 5000 -> vocab size(top 5000 words)
    # 64 -> each word will become a list of 64 numbers
    # Simple words-> Converted words into codes so that computer can understand them

    # Step 3: Understanding the sequence (GRU layer)
    tf.keras.layers.GRU(128),
    # This is the memory part of the model/brain
    # Reads the sentence word by word
    # Remembers only the important words
    # Forgets the useless words
    # Example: 'I did not like the food at APD' -> GRU focusses on 'not like' -> negative meaning
    # 128 -> Size of the memory (how much your model can remember)

    # Step 4: Final Decision
    tf.keras.layers.Dense(1,activation='sigmoid')
    # This output layer gives the final answer
    # Output -> A number between 0 and 1.
    # Example: 0.9 -> Positive response/ positive sentiment and 0.1 -> Negative response/negative sentiment
    # sigmoid: Converts output into probability
])

# Step 5: Preparing the model
model.compile(optimizer='adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
# This tells the model how to learn
# optimizer = 'adam', helps the model to improve itself
# loss = 'binary_crossentropy', measures how wrong the prediction is
# accuracy = How often your model is correct

model.summary()
```

```python
# # Difference between GRU and LSTM
# # Basic idea
# 1. LSTM (Long Short-Term memory): Designed to remmeber information for long sequences using a complex memory system.
# 2. GRU (Gated recurrent unit): A simplified version of LSTM with fewer gates and faster implementation.

# Architecture differences
# LSTM has 3 gates and 2 states which are
# 1. input gate -> decides what to add.
# 2. output gate -> Decides what to output
# 3. forget gate -> Decides what to remove
# Cell state: Long term memory
# Hidden state: Short term memory

# GRU has only 2 gates:
# 1. Update gate: combines forget + input
# 2. Reset gate -> controls past information


# Complexity
# LSTM is more complex because of 3 gates and 2 states

# Performance:
# LSTM: Better for long sequences, more control over memory, slower
# GRU: faster to train, works well with small data
```
