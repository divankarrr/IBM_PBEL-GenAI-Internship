## ✂️ Sentence vs Word Tokenization

### 📌 Sentence Tokenization
Breaking text into sentences.

**Example:**
"AI is powerful. It is everywhere."

→ ["AI is powerful.", "It is everywhere."]

### 📌 Word Tokenization
Breaking sentence into words.

→ ["AI", "is", "powerful"]

### 🔥 Difference
| Sentence Tokenization | Word Tokenization |
|----------------------|------------------|
| Splits into sentences | Splits into words |
| Higher level         | Lower level      |

## 🌱 Stemming vs Lemmatization

### 📌 Stemming
Cuts words to root form (may be incorrect)

running → run  
studies → studi ❌

### 📌 Lemmatization
Returns meaningful root word

running → run  
studies → study ✅

### 🔥 Difference
| Stemming | Lemmatization |
|----------|--------------|
| Fast     | Slower       |
| Less accurate | More accurate |
| No context | Uses context |


```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

# words = ["running", "studies", "Intelligence", "porter", "Finder", "keeper"]

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

for w in words:
    print("Stem:", stemmer.stem(w), "| Lemma:", lemmatizer.lemmatize(w))
```


## 📊 Count Vectorizer vs TF-IDF

### 📌 Count Vectorizer
Counts word frequency

"NLP is powerful powerful"
→ powerful = 2

### 📌 TF-IDF (Term Frequency-Inverse Document Frequency)

Components of TF-IDF:

1. Term Frequency (TF): Measures how frequently a word appears in a specific document.

2. Inverse Document Frequency (IDF): Measures how important a word is across all documents in the corpus. It reduces the weight of common words (e.g., "the", "a") and increases the weight of rare words.

Means:

- Weights important words higher

- Rare words → higher importance

### 🔥 Difference
| Count Vectorizer | TF-IDF |
|----------------|--------|
| Raw frequency  | Weighted importance |
| Simple         | More informative |


```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
corpus = sent
corpus = ["NLP is powerful", "NLP is easy"]

cv = CountVectorizer()
tfidf = TfidfVectorizer()
print("Count:", cv.fit_transform(corpus).toarray())
print("TF-IDF:", tfidf.fit_transform(corpus).toarray())
print(cv.get_feature_names_out())
print(tfidf.get_feature_names_out())

# easy = 0
# is = 1
# NLP = 2
# powerful = 3
```


## 🔗 N-grams


## 🧠 POS Tagging

### 📌 Definition
Identifies grammatical role of words

Example:
"I run fast"

I → Pronoun  
run → Verb  
fast → Adverb


```python
import nltk
# nltk.download('averaged_perceptron_tagger')

# words = ["I", "run", "fast"]
print(nltk.pos_tag(words))
```

## 1. Basic Text Generation (Default)

📌 What’s happening:
- Uses default model (usually GPT-2)
- Generates continuation of input text

```python
from transformers import pipeline

generator = pipeline("text-generation")

print(generator("AI will", max_length=20))
```

## 2. Controlled Output (Temperature & Sampling)

📌 Key Parameters:
temperature → randomness (low = deterministic, high = creative)
top_k → limits vocabulary choices
top_p → nucleus sampling

👉 Use this when you want better quality + less randomness

```python
from transformers import pipeline

generator = pipeline("text-generation")

print(generator(
    "AI will",
    max_length=20,
    temperature=0.7,
    top_k=50,
    top_p=0.9
))
```

## 3. Using a Specific Model

📌 Why this matters:
Different models = different performance
gpt2-medium > default GPT-2 in quality

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="gpt2-medium"
)

print(generator("AI will change the world because", max_length=30))
```

## 🔁 4. Multiple Outputs (n variations)

📌 Use case:
Brainstorming ideas
Generating multiple responses
Creative applications

```python
from transformers import pipeline

generator = pipeline("text-generation")

outputs = generator(
    "AI will",
    max_length=20,
    num_return_sequences=3
)

for i, o in enumerate(outputs):
    print(f"Output {i+1}: {o['generated_text']}")
```
