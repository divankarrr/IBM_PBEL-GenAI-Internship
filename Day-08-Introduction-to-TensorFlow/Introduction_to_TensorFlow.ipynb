## What is TensorFlow?

**TensorFlow** is an **open-source machine learning and deep learning framework** developed by **Google Brain** in 2015.  
It provides a flexible ecosystem of tools, libraries, and community resources that make it easier to build and deploy machine learning applications.  

TensorFlow enables developers to perform **numerical computation** using **data flow graphs**, where nodes represent mathematical operations and edges represent multidimensional data arrays called *tensors*.  
It is widely used for building and training **machine learning models**, especially **neural networks**.

TensorFlow can run computations on various hardware platforms such as **CPUs**, **GPUs**, and **TPUs**, making it highly scalable from a single laptop to large-scale distributed systems.

---

### Key Uses of TensorFlow

- **Machine Learning:**  
  Used for building and training supervised and unsupervised models like regression, classification, and clustering.

- **Deep Learning:**  
  Supports a variety of neural network architectures such as Convolutional Neural Networks (CNNs), Recurrent Neural Networks (RNNs), and Transformers.

- **Numerical Computation:**  
  TensorFlow can perform efficient linear algebra, optimization, and matrix operations using tensors.

- **Research and Production:**  
  TensorFlow is designed for both research prototyping and production deployment. It supports model serving, mobile deployment (TensorFlow Lite), and web applications (TensorFlow.js).

- **Cross-Platform Flexibility:**  
  Models trained on one platform can be deployed on different environments with minimal effort.

---

### TensorFlow Architecture

TensorFlow’s architecture consists of three main components:

1. **Client**  
   - The client is where users define and configure computational tasks.  
   - It typically uses Python APIs (though C++, Java, and Swift APIs also exist).  
   - The client constructs a *computational graph* that describes the flow of data and operations.

2. **TensorFlow Runtime (Master and Worker Services)**  
   - The runtime executes the computational graph created by the client.  
   - It manages resources such as CPUs, GPUs, and TPUs.  
   - TensorFlow can distribute the graph across multiple devices and machines to achieve parallel and efficient computation.  
   - The master service coordinates the execution, while worker services handle computation on individual devices.

3. **Execution Model**  
   TensorFlow supports two main execution modes:
   - **Graph Execution (TensorFlow 1.x):**  
     The computation is first defined as a graph and then executed within a `Session`. This allows optimization before execution and efficient distributed processing.
   - **Eager Execution (TensorFlow 2.x):**  
     Operations are executed immediately as they are called, which makes TensorFlow more intuitive and easier to debug. This is the default mode in TensorFlow 2.x.

## What is Keras?

**Keras** is an **open-source high-level neural network API** written in Python.  
It was developed by **François Chollet** in 2015 and is now **integrated into TensorFlow** as its official high-level API (`tf.keras`).  

Keras provides a user-friendly, modular, and extensible interface for building and training deep learning models.  
It abstracts much of the complexity of TensorFlow, allowing developers to create neural networks with just a few lines of code.

Unlike low-level frameworks that require manually defining tensors and operations, Keras focuses on **simplicity and ease of experimentation**, making it ideal for beginners and researchers who want to prototype models quickly.

---

### Key Uses of Keras

- **Deep Learning Model Development:**  
  Used to design and train various neural networks, including CNNs, RNNs, autoencoders, and Transformers.

- **Prototyping and Research:**  
  Provides an intuitive, modular API that allows rapid model development and testing of new ideas.

- **Transfer Learning:**  
  Offers access to pre-trained models (like VGG, ResNet, and Inception) that can be fine-tuned for new tasks.

- **Production Deployment:**  
  Integrated with TensorFlow for deploying trained models to mobile, web, or cloud environments using TensorFlow Lite and TensorFlow Serving.

- **Educational Use:**  
  Keras’ readability and simplicity make it a great tool for teaching deep learning concepts.

---

### Keras Architecture

Keras follows a **modular architecture**, which means that each component of a neural network (layers, models, optimizers, losses, etc.) can be combined freely to build new models.

Keras consists of the following key components:

1. **Models**  
   - A **model** defines how layers are connected.  
   - The most commonly used is the **Sequential API**, which allows stacking layers linearly.  
   - For more complex networks, the **Functional API** is used to build directed acyclic graphs of layers.


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(64, activation='relu', input_shape=(100,)),
    Dense(10, activation='softmax')
])
```

# What are Tensors?

Tensors are **multi-dimensional arrays** used by TensorFlow to represent data.  
They can hold numerical values of various types (`int`, `float`, etc.) and are the core data structure in TensorFlow.

---

### Tensor Ranks
The **rank** of a tensor = number of dimensions (axes) it has.

| Rank | Example | Shape | Description |
|------|----------|--------|--------------|
| 0 | `tf.constant(5)` | `()` | Scalar (single value) |
| 1 | `tf.constant([1, 2, 3])` | `(3,)` | Vector |
| 2 | `tf.constant([[1, 2], [3, 4]])` | `(2, 2)` | Matrix |
| 3 | 3D tensor | `(2, 2, 2)` | Cube / stack of matrices |
| n | n-D tensor | `(d1, d2, …, dn)` | Higher-dimensional data |

# TensorFlow — Data Types (tf.dtypes)

TensorFlow uses tensors (multidimensional arrays) where each tensor has a data type (dtype). The tf.dtypes module defines all supported types. Choosing the correct dtype affects memory, numeric precision, speed, and which ops are allowed.

Common categories:

* Boolean: logical values.

* Integer: signed / unsigned, various bit widths.

* Floating point: float16 / bfloat16 / float32 / float64.

* Complex: pairs of floats for real/imag parts.

* String: byte sequences.

* Resource / Variant: internal and opaque types (e.g., for datasets, resource handles).

* Quantized: specialized integer types used by quantized kernels.

| Category | Data Type | Alias / Example | NumPy Equivalent | Bits |
|-----------|------------|----------------|------------------|------|
| **Boolean** | `tf.bool` | `tf.constant(True)` | `np.bool_` | 1 |
| **Signed Int** | `tf.int8`, `tf.int16`, `tf.int32`, `tf.int64` | `tf.constant(5, tf.int32)` | `np.int8`–`np.int64` | 8–64 |
| **Unsigned Int** | `tf.uint8` | `tf.constant(255, tf.uint8)` | `np.uint8` | 8 |
| **Float** | `tf.float16`, `tf.bfloat16`, `tf.float32`, `tf.float64` | `tf.constant(1.5, tf.float32)` | `np.float16`–`np.float64` | 16–64 |
| **Complex** | `tf.complex64`, `tf.complex128` | `tf.constant(1+2j)` | `np.complex64`, `np.complex128` | 64/128 |
| **String** | `tf.string` | `tf.constant("hello")` | `np.object_` | variable |
| **Quantized** | `tf.qint8`, `tf.quint8`, `tf.qint32` | – | – | 8–32 |
| **Special** | `tf.resource`, `tf.variant` | internal | – | – |

```python
a = tf.constant([1, 2, 3], dtype=tf.int32)
b = tf.constant([1.2, 3.4], dtype=tf.float32)
s = tf.constant(["A", "B"], dtype=tf.string)

print(a.dtype, b.dtype, s.dtype)

# Casting
x = tf.cast(a, tf.float32)
y = tf.cast(b, tf.int32)
print(x.dtype, y.dtype)
```

# Computational Graphs

TensorFlow builds and executes operations using computational graphs.

### What is a Computational Graph?
A **computational graph** is a series of operations (nodes) connected by data (edges).  
Each node performs a specific computation (like addition or multiplication), and edges carry the output (tensors) between nodes.

In TensorFlow:
- **Nodes** = operations (e.g., `add`, `multiply`)
- **Edges** = tensors (data flowing between operations)

TensorFlow first **builds the graph** (defines what to compute), and then **executes** it (runs the actual computation).

### Benefits of Graphs
- Faster execution through **parallel processing**
- Can run on multiple devices (CPU, GPU, TPU)
- Optimizes memory and computation automatically
- Easy to **save**, **export**, and **reuse**

# What is a Session?

A **Session** is an environment used to **execute the computational graph**.  
It allocates system resources (like CPU, GPU, and memory) and runs the defined TensorFlow operations.

### Key Points

- You must **build the graph first**, then **execute it inside a Session**.  
- Use `sess.run()` to fetch and compute results.  
- In **TensorFlow 2.x**, sessions are **no longer required** because of **eager execution** (operations run immediately as written).


```python
import tensorflow as tf

# TensorFlow 2.x style (eager execution)

# Operations are executed immediately
x = tf.constant(3)
y = tf.constant(4)
z = x * y + 2

print(z) # Output: tf.Tensor(14, shape=(), dtype=int32)
```

```python
# Example: TensorFlow 1.x style
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

# Build the graph
x = tf.constant(3)
y = tf.constant(4)
z = x * y + 2

# Run the graph in a session
with tf.Session() as sess:
    result = sess.run(z)
    print(result)  # Output: 14
```
